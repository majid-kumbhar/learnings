```python
x_sin = np.arange(0,2*np.pi,0.1)
y_sin = np.sin(x_sin)

print(y_sin)
```

```python
plt.figure(figsize = (3,3))
plt.plot(x_sin, y_sin)
plt.title("Sin Curve")
plt.show()
```

```python
x_cos = np.arange(0,2*np.pi, 0.1)
y_cos = np.cos(x_cos)

plt.figure(figsize = (3,3))
plt.plot(x_cos, y_cos)
plt.title('Cos Curve')
plt.show()
```

```python
x_tan = np.arange(0,2*np.pi, 0.1)
y_tan = np.tan(x_tan)

plt.figure(figsize = (3,3))
plt.plot(x_tan, y_tan)
plt.title('tan Curve')
plt.show()
```

```python
# loc() Location  function in pandas
sample_data = pd.read_csv(r"C:\Users\reals\Downloads\sample2.csv",index_col = ['Roll No.'])
sample_data.head() # Data will be started from Roll Number wise not from '0' index
sample_data.loc[1] # It will give us all the value of given index line wise 
sample_data.loc[[5,6,7,8]] # Will give onlly the given index's data 
sample_data.loc[5,'Physics'] # Gives only fifth index's phyics data
sample_data.loc[1:5,'Physics'] # Gives 1 to 5 index's physics data
sample_data.loc[sample_data['Physics']<50] # Gives data of physics whose roll no is between 1 tp 49
```

```python
# Group by() function in pandas
grouped_data = sample_data.groupby(['Branch'])
grouped_data.groups # makes a group of each branch with roll numbers available in each group
branch_section = sample_data.groupby(['Branch', 'Section'])
branch_section.groups # Makes a group of each branch and sections and add roll numbers that are present in same branch and section
for group, data in branch_section: # Running a loop to get the whole data line by line for each branch & section data 
    print(group) 
    print(data)



```
