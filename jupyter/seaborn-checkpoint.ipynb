# line plot in seaborn

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np 
```

```python
roll_num=range(1,16)
marks=[23,21,2,3,2,12,67,33,56,15,35,67,34,3,32]
len(marks)
len(roll_num)
```

```python
df=pd.DataFrame({'roll_num':roll_num,'marks':marks})
df
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='roll_num',y='marks' ,data=df)

plt.title("student mark")
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='roll_num',y='marks' ,data=df)

plt.title("student mark")
```

```python
df1 = sns.load_dataset('penguins').head(20)
df1
```

```python
# plt.figure(figsize = (12,6))
# style = "sex" , changes the plain line to dotted line.
 # size = 50  changes the size of the graph.
# palette = "Any color name here" , palette = "Anything" shows the colors that can be used as a parameter
# markers = ['o','>'] adds markers in the plot
# dashes = false # removes the dots from the style dots 
# legend = False # Removes the legend 
sns.lineplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df1 , hue = 'sex',style = 'sex' , palette = "Accent", markers =['o','>'], dashes = False, legend=False)
plt.grid() # adds grids in the table
plt.title('Majid Bhaiiiii') # adss title of majid bhaiiiiii'
plt.show()
```
