```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./datasets/Customer Call List.csv")
df
# first step look at the data clearly 
#standrize it 
```

```python
# Removing Duplicate 
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# Standrizing and cleaning the messy last Name column by using str.strip() method
df['Last_Name'] = df['Last_Name'].str.strip('123_./')
df
```

```python
# standrize & Clears the phone number column using the regex method 
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex = True)
df
```

```python
# using the lambda function to drive a loop through the phone number column to assign a '-' after every three numbers
# First we convert the column value into string to perfomr the lambda function according to our need 
# df['Phone_Number']= df['Phone_Number'].apply(lambda x: str(x)) # converts it to string 
# df['Phone_Number'].info()
df['Phone_Number']= df['Phone_Number'].apply(lambda x: x[0:3]+ '-' + x[3:6] + '-' + x[6:10]) #this lambda add '-' after every number in the row 
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
```

```python
df['Phone_Number']
```

```python
df[['Street_Address',"State","Zip_Code"]]=df['Address'].str.split(',',n=2, expand =True)
# splits the Address column value into three different columns
df
```

```python
df = df.drop(columns='Address')
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('N','No')
df['Paying Customer']=df['Paying Customer'].str.replace('Noo','No')
df['Paying Customer']=df['Paying Customer'].str.replace('Y','Yes')
df['Paying Customer']=df['Paying Customer'].str.replace('Yeses','Yes')
```

```python
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Noo','No')
df['Do_Not_Contact']
```
